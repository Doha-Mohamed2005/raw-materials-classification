```python
import kagglehub

path = kagglehub.dataset_download("haaroonafroz/material-dataset-new")
print("Path to dataset:", path)
```

```python
import os

for item in os.listdir(path):
    print(item)
```

```python
import shutil

def move_dataset(source, destination):
    shutil.move(source, destination)
    print(f"Dataset moved from {source} to {destination}")

move_dataset(path, r"Downloads")
```
